```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import plot_histogram
import math
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit_aer import AerSimulator, StatevectorSimulator, QasmSimulator
# from qiskit.visualization import plot_histogram

%matplotlib inline
```

```python
circuit = QuantumCircuit(3,3)

circuit.x(0)
circuit.barrier()
circuit.h(1)
circuit.cx(1,2)
circuit.barrier()
draw = circuit.draw('mpl')
```

```python
circuit.cx(0,1)
circuit.h(0)
circuit.barrier()
draw = circuit.draw('mpl')
```

```python
circuit.measure([0,1],[0,1])
circuit.barrier()

draw = circuit.draw('mpl')
```

```python
circuit.cx(1,2)
circuit.cz(0,2)
# circuit.barrier()
circuit.measure([2],[2])
draw = circuit.draw('mpl')
```

```python
from qiskit_aer import Aer

backend = Aer.get_backend('qasm_simulator')
result = backend.run(transpile(circuit, backend=backend)).result()
hist = plot_histogram(result.get_counts(circuit))
```
