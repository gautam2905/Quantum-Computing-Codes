```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import plot_histogram
import math
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit_aer import AerSimulator, StatevectorSimulator, QasmSimulator
# from qiskit.visualization import plot_histogram

%matplotlib inline
```

```python
# For execution
# simulator = QasmSimulator()
# backend = StatevectorSimulator()

def run_on_simulator(circuit):
    backend = Aer.get_backend('statevector_simulator')
    result = backend.run(transpile(circuit, backend)).result()
    psi = result.get_statevector()

    back = Aer.get_backend('qasm_simulator')
    num_bits = circuit.num_qubits
    circuit.measure([i for i in range(num_bits)] , [i for i in range(num_bits)] )
    qasm_job = back.run(transpile(circuit , back)).result()
    counts = qasm_job.get_counts()

    return psi, counts
```
